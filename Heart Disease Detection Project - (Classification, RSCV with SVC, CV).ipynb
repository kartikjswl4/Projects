# Heart Disease Detection Project


Problem Statement:

This database contains 14 attributes . In particular, the Cleveland database is the only one that has been used by ML researchers to this date. 
The "goal" field refers to the presence of heart disease in the patient. It is integer valued from 0 (no presence) to 4. Experiments with the Cleveland database have concentrated on simply attempting to distinguish presence (values 1,2,3,4) from absence (value 0).



Only 14 attributes used:
1. (age)
2. sex     // (1 = male; 0 = female)
3. cp      // chest pain type
4. trestbps// resting blood pressure (in mm Hg on admission to the hospital)
5. chol    // serum cholestoral in mg/dl
6. fbs     // (fasting blood sugar &gt; 120 mg/dl) (1 = true; 0 = false)
7. restecg // resting electrocardiographic results
8. thalach // maximum heart rate achieved
9. exang   // exercise induced angina (1 = yes; 0 = no)
10. oldpeak// ST depression induced by exercise relative to rest
11. slope  // the slope of the peak exercise ST segment
12. ca     // number of major vessels (0-3) colored by flourosopy
13. thal   //  0 = normal; 1 = fixed defect; 2 = reversable defect
14. target

```python
X = df1.drop(['target'], axis = 1)
Y = df1['target']
```

```python
#Transforming data to remove skewness
from sklearn.preprocessing import power_transform
X = power_transform(X, method='yeo-johnson')
X
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
X
```

# Train test split 

```python
df['target'].value_counts()
```

```python
Y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 34)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
#Logistic Regression

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, Y_train)
pred = lr.predict(X_test)

print(confusion_matrix(Y_test, pred))
print(classification_report(Y_test, pred))
print(accuracy_score(Y_test, pred))
cvs = cross_val_score(lr, X, Y, cv=None)
cvs.mean()
```

```python
#RandomForest

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,Y_train)
pred = rf.predict(X_test)

print(confusion_matrix(Y_test, pred))
print(classification_report(Y_test, pred))
print(accuracy_score(Y_test, pred))
cvs = cross_val_score(rf, X, Y, cv=None)
cvs.mean()
```

```python
#DescisonTree

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train,Y_train)
pred = dt.predict(X_test)

print(confusion_matrix(Y_test, pred))
print(classification_report(Y_test, pred))
print(accuracy_score(Y_test, pred))
cvs = cross_val_score(dt, X, Y, cv=None)
cvs.mean()
```

```python
#SVC

from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, Y_train)
pred = svc.predict(X_test)

print(confusion_matrix(Y_test, pred))
print(classification_report(Y_test, pred))
print(accuracy_score(Y_test, pred))
cvs = cross_val_score(svc, X, Y, cv=None)
cvs.mean()
```

```python
# KNN

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train, Y_train)
pred = knn.predict(X_test)

print(confusion_matrix(Y_test, pred))
print(classification_report(Y_test, pred))
print(accuracy_score(Y_test, pred))
cvs = cross_val_score(knn, X, Y, cv=None)
cvs.mean()
```

Suppose these 2 scenarios -
1. Model says a person has cancer when in real he hasn't. (False positive)
2. Model says a person doesn't have cancer but he has. (False negative)
Which is more dangerous in this case? 
- The 2nd one right! As wrong detection will make that person live carefree but in real he'll be getting more ill and may also die.

In this case the we will select the model with the lowest False Negative considering the fact that the proportion of actual positives are correctly classified i.e model tells a person has cancer and he actually does.
The less the no. of false negatives the more the model is perfect in this case.

# Hyperparameter Tuning 

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
C=range(5, 280, 55)
kernel= ['linear', 'poly', 'rbf', 'sigmoid']
degree=[1,2,3,4,5,6]
gamma=['scale', 'auto']
cache_size=[10, 20, 30]
decision_function_shape=['ovo', 'ovr']

random_grid = {'C': C,
               'kernel' : kernel,
               'degree' : degree,
               'gamma' : gamma,
               'cache_size' : cache_size,
               'decision_function_shape' : decision_function_shape}
random_grid
```

```python
svc = SVC()
rscv = RandomizedSearchCV(estimator = svc,
                         param_distributions = random_grid,
                          cv = 10,
                          random_state = 25,
                         verbose = 2,
                         n_jobs = -1)

rscv.fit(X_train, Y_train)
rscv
```

```python
rscv.best_params_
```

```python
rscv.best_estimator_
```

```python
best_rscv = rscv.best_estimator_
pred = best_rscv.predict(X_test)

print(confusion_matrix(Y_test, pred))
print(classification_report(Y_test, pred))
print(accuracy_score(Y_test, pred))

cvs = cross_val_score(rscv, X, Y, cv=10)
cvs.mean()
```
