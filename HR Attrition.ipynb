# Comparative Analysis of Department and Average Total Working years

```python
import math
```

```python
#Creating a empty list to store the values of individual result of department
final=[]
depts=['Research & Development','Sales','Human Resources']

#figure definition
figure(figsize=(16,8),dpi=80)
ax=plt.axes()
ax.set_facecolor('Black')

#Department of Research & Development
cc=a[a['Department']=='Research & Development']
Average_Total_Working_years_ResearchDevelopment=((cc['TotalWorkingYears'].sum())/len(a['TotalWorkingYears']))
final.append(Average_Total_Working_years_ResearchDevelopment)

#Department of Sales
aa=a[a['Department']=='Sales']
Average_Total_Working_years_sales=((aa['TotalWorkingYears'].sum())/len(a['TotalWorkingYears']))
final.append(Average_Total_Working_years_sales)


#Department of Human Resources
bb=a[a['Department']=='Human Resources']
Average_Total_Working_years_HumanResources=((bb['TotalWorkingYears'].sum())/len(aa['TotalWorkingYears']))
final.append(Average_Total_Working_years_HumanResources)

final
plt.plot(depts,final,color='white')
plt.bar(depts,final,color='blue',width=0.01)
plt.scatter(depts,final,color='red',s=150)
```

1. From the above pictorial representation it is clear that the employees of Research and Development department stays with the company for the long time.

2. The second position with the sales department and then last it comes with Human Resource department

Finally the Attrition rate will be higher in Research and Development sector
