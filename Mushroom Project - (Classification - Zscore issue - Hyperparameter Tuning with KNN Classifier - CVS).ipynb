

# Mushroom Project

Problem Statement:
This data set includes descriptions of hypothetical samples corresponding to 23 species of gilled mushrooms in the Agaricus and Lepiota Family (pp. 500-525). 
Each species is identified as definitely edible, definitely poisonous, or of unknown edibility and not recommended. 
This latter class was combined with the poisonous one. 
The Guide clearly states that there is no simple rule for determining the edibility of a mushroom; no rule like ``leaflets three, let it be'' for Poisonous Oak and Ivy.

Thus, in this project, you have to predict the "Class" column. 

Downlaod Files:
https://github.com/dsrscientist/dataset1/blob/master/mushrooms.csv

# Train Test Split

```python
Y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 34)
```

```python
#Logistic Regression

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, Y_train)
pred = lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print(confusion_matrix(Y_test, pred))
print(classification_report(Y_test, pred))
print(accuracy_score(Y_test, pred))
```

```python
#RandomForest

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,Y_train)
pred = rf.predict(X_test)
```

```python
print(confusion_matrix(Y_test, pred))
print(classification_report(Y_test, pred))
print(accuracy_score(Y_test, pred))
```

```python
#DescisonTree

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train,Y_train)
pred = dt.predict(X_test)
```

```python
print(confusion_matrix(Y_test, pred))
print(classification_report(Y_test, pred))
print(accuracy_score(Y_test, pred))
```

```python
#SVC

from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, Y_train)
pred = svc.predict(X_test)
```

```python
print(confusion_matrix(Y_test, pred))
print(classification_report(Y_test, pred))
print(accuracy_score(Y_test, pred))
```

```python
# KNN

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train, Y_train)
pred = knn.predict(X_test)
```

```python
print(confusion_matrix(Y_test, pred))
print(classification_report(Y_test, pred))
print(accuracy_score(Y_test, pred))
```

# Hyperparameter Tuning 

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_neighbors    = range(1, 8000, 150)
weights        = ['uniform', 'distance']
algorithm      = ['auto', 'ball_tree', 'kd_tree', 'brute']
leaf_size      = range(1, 7500, 200)
metric         = ['minkowski']

random_grid = {'n_neighbors' : n_neighbors,
               'weights'     : weights,
               'algorithm'   : algorithm,
               'leaf_size'   : leaf_size,
               'metric'      : metric}

print(random_grid)
```

```python
knn = KNeighborsClassifier()
rscv = RandomizedSearchCV(estimator = knn,
                         param_distributions = random_grid,
                         cv = 10,
                         random_state=100,
                         verbose = 3,
                         n_jobs = -1)

rscv.fit(X_train, Y_train)
rscv
```

```python
rscv.best_params_
```

```python
rscv.best_estimator_
```

```python
best_rscv = rscv.best_estimator_
```

```python
pred = best_rscv.predict(X_test)
```

```python
print(confusion_matrix(Y_test, pred))
print(classification_report(Y_test, pred))
print(accuracy_score(Y_test, pred))
```

# Stratified K Fold

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits = 10)
skf.split(X,Y)
```

```python
accuracy= []

skf = StratifiedKFold(n_splits=10, random_state = None)
skf.get_n_splits(X,Y)

#X is the feature set and Y is the target
#Putiing train_index and test_index of X and Y

for train_index, test_index in skf.split(X,Y):
    print('Train:', train_index, 'Validation:', test_index)
    X1_train, X1_test = X.iloc[train_index], X.iloc[test_index]
    Y1_train, Y1_test = Y.iloc[train_index], Y.iloc[test_index]
    
rscv.fit(X1_train, Y1_train)
pred = rscv.predict(X1_test)
score = accuracy_score(pred, Y1_test)
accuracy.append(score)
```
